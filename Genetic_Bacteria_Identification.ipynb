Bakteriyel antibiyotik direncinin hızla yayılmasının önüne geçmek, hayat kurtarıcı olabilir. Bu doğrultuda, bakterilerin genomik yapılarının analiz edilmesi ve hangi antibiyotiklere dirençli olduklarının belirlenmesi, tedavi sürecini hızlandırmak ve doğru tedavi seçenekleri sunmak adına kritik bir adımdır. **Eksik Değerler ve Veri Temizleme:**

Veri setinde eksik değerler (null) kontrol ediliyor ve eksik veriler belirli bir değerle dolduruluyor.
Veri setinde gereksiz olan sütunlar (row_id) kaldırılıyor.

```python
def reduce_mem_usage(props):
    # Başlangıç bellek kullanımını hesaplama ve yazdırma
    start_mem_usg = props.memory_usage().sum() / 1024**2
    print("Properties dataframe'inin bellek kullanımı:", start_mem_usg, " MB")

    NAlist = []  # Eksik değerlerin doldurulduğu sütunları takip eder.

    for col in props.columns:
        if props[col].dtype != object:  # String türlerini hariç tutma
            # Mevcut sütun tipi hakkında bilgi yazdırma
            print("******************************")
            print("Sütun: ", col)
            print("Öncesindeki dtype: ", props[col].dtype)

            # Integer tipi için değişkenler oluştur, maksimum ve minimum değerleri alma
            IsInt = False
            mx = props[col].max()
            mn = props[col].min()


            if not np.isfinite(props[col]).all():
                NAlist.append(col)
                props[col].fillna(mn-1, inplace=True)

            # Sütunun integer'a dönüştürülüp dönüştürülemeyeceğini test etme
            asint = props[col].fillna(0).astype(np.int64)
            result = (props[col] - asint)
            result = result.sum()
            if result > -0.01 and result < 0.01:
                IsInt = True

            # Integer veya unsigned Integer türleri oluşturma
            if IsInt:
                if mn >= 0:
                    if mx < 255:
                        props[col] = props[col].astype(np.uint8)
                    elif mx < 65535:
                        props[col] = props[col].astype(np.uint16)
                    elif mx < 4294967295:
                        props[col] = props[col].astype(np.uint32)
                    else:
                        props[col] = props[col].astype(np.uint64)
                else:
                    if mn > np.iinfo(np.int8).min and mx < np.iinfo(np.int8).max:
                        props[col] = props[col].astype(np.int8)
                    elif mn > np.iinfo(np.int16).min and mx < np.iinfo(np.int16).max:
                        props[col] = props[col].astype(np.int16)
                    elif mn > np.iinfo(np.int32).min and mx < np.iinfo(np.int32).max:
                        props[col] = props[col].astype(np.int32)
                    elif mn > np.iinfo(np.int64).min and mx < np.iinfo(np.int64).max:
                        props[col] = props[col].astype(np.int64)

            # Float türlerini 32 bit olarak dönüştürme
            else:
                props[col] = props[col].astype(np.float32)

            # Yeni sütun tipini yazdırma
            print("Sonraki dtype: ", props[col].dtype)
            print("******************************")

    # Sonuçları yazdırma
    print("___İŞLEM SONRASI BELLEK KULLANIMI:___")
    mem_usg = props.memory_usage().sum() / 1024**2
    print("Bellek kullanımı: ", mem_usg, " MB")
    print("Bu, başlangıç boyutunun ", 100 * mem_usg / start_mem_usg, "%'i kadar.")
    return props, NAlist
```

**Bellek Kullanımını Azaltma:**

reduce_mem_usage fonksiyonu ile veri çerçevesindeki her sütunun veri tipi optimize edilerek bellek kullanımı azaltılıyor.

```python
import numpy as np
train, NAlist = reduce_mem_usage(train)
print("_________________")
print("")
print("Uyarı: Aşağıdaki sütunlar, eksik değerler 'df['column_name'].min() -1' ile doldurulmuştur: ")
print("_________________")
print("")
print(NAlist)
# 'reduce_mem_usage' fonksiyonu, veri çerçevesindeki bellek kullanımını azaltmak için her sütunun veri tipini optimize eder.
# Bu fonksiyon, daha küçük veri tiplerine dönüştürerek bellek kullanımını azaltır ve aynı zamanda eksik değerler varsa bunları
# belirli bir değeri (min() - 1) ile doldurur.
# 'NAlist' listesi, eksik verilerin bulunduğu sütunları içerir.
# Bu işlem, eksik değerlerin yerine uygun bir değer (min() - 1) yerleştirilerek, modelin veri üzerinde işlem yapabilmesi sağlanır.
# Bu işlemden sonra, eksik değerlerin hangi sütunlarda bulunduğunu ve bu sütunların hangi şekilde doldurulduğunu ekrana yazdırır.
```
